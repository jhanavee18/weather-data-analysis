```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("daily_weather.csv")
data.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['air_temp_9am'])
plt.title("Air temperature at 9 AM over time")
plt.xlabel("Day number")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(data['rain_accumulation_9am'])
plt.title("Rain accumulation at 9 AM over time")
plt.xlabel("Day number")
plt.ylabel("Rainfall accumulation")
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(data['air_temp_9am'], data['relative_humidity_9am'])
plt.title("Temperature VS Humidity at 9 AM")
plt.xlabel("Air temperature at 9 AM")
plt.ylabel("Relative humidity at 9 AM")
plt.show()
```

```python
print("Average temperature at 9AM:", data['air_temp_9am'].mean())
print("Maximum temperature at 9 AM:", data['air_temp_9am'].max())
print("Minimum temperature at 9 AM:", data['air_temp_9am'].min())
print("Average Rainfall at 9 AM:", data['rain_accumulation_9am'].mean())
```

##project insights

1. The air temperature at 9 AM shows variation over time, indicating chnaging weather patterns.
2. The rainfall accumulation is mostly low, and some occasional spikes indicating heavy rainfall.
3. The scatter plot between temperature and humidity shows an inverse relationship i.e. as temperature rises humidity decreases.
4. The average temperature gives an overall idea of the climate, while the maximum and minimum values shows the extremes.
5. This dataset helps understand daily weather behaviour and can be used for future prediction models.
